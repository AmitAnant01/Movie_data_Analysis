## Now Categorizing Vote_Average column

We would cut the Vote_Average values and make 4 categories popular, average, below_avg and not popular to descrbe it more categorize_col() function provided above

```python

def catigorize_col(data,col,label):
    edges = [data[col].describe()['min'],
             data[col].describe()['25%'],
             data[col].describe()['50%'],
             data[col].describe()['75%'],
             data[col].describe()['max']]

    data[col] = pd.cut(data[col],edges, labels = label, duplicates = 'drop')
    return data
             
```

```python
label= ['not_popular','below_average','average','popular']
catigorize_col(data,'Vote_Average', label)
data['Vote_Average'].unique()
```
